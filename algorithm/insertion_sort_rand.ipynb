```python
#조건 : 오름차순/랜덤요소 정렬

import random
import time
import datetime
import matplotlib.pyplot as plt

# 삽입 정렬 함수
def insertion_sort(list1):
    global countNum
    global countLoopNum
    countNum,countLoopNum = 0,0

    for j in range(1,len(list1)):
        countNum += 1
        key = list1[j]
        i = j-1
        while i >= 0 and list1[i] > key:
            list1[i+1] = list1[i]
            i -= 1
            list1[i+1] = key
            countLoopNum += 1

# 1부터 100까지 숫자를  10000개~20000개 추출하여 리스트에 넣는 함수
def random_make_num(list2):
    global rand_num
    rand_num = random.randrange(10000,20000)
    for i in range(rand_num):
        randNum = int(random.randrange(1,100))
        list2.append(randNum)
```

```python
# 초기화된 리스트 생성
randList1 = []
randList2 = []
randList3 = []
```

```python
# 각 케이스에 대하여 삽입 정렬, 요소 개수 n 출력 및 While Loop test 실행 횟수 출력
# countNum에 +1을 하는 이유는 for loop test는 loop body보다 한번 더 실행되며, 이것이 요소의 개수인 n이 된다.
# countLoopNum에 rand_num-1을 하는 이유는 while loop test는 while loop body보다 (n-1)번 더 실행되기 때문이다.

start1 = time.time()

random_make_num(randList1)

print("Random input case list_sort before: ",randList1)
insertion_sort(randList1)
print("Random input case list_ sort after: ",randList1)
print("For Loop test number of times: ",countNum+1)
randList1_countNum = countNum+1
countNum = 0
print("Inner While Loop test number of times: ",countLoopNum+rand_num-1)
randList1_countLoopNum = countLoopNum+rand_num-1
rand_num = 0
countLoopNum = 0

sec1 = time.time()-start1
```

```python
start2 = time.time()

random_make_num(randList2)

print("Random input list_sort before: ",randList2)
insertion_sort(randList2)
print("Random input list_ sort after: ",randList2)
print("For Loop test number of times: ",countNum+1)
randList2_countNum = countNum+1
countNum = 0
print("Inner While Loop test number of times: ",countLoopNum+rand_num-1)
randList2_countLoopNum = countLoopNum+rand_num-1
rand_num = 0
countLoopNum = 0

sec2 = time.time()-start2
```

```python
start3 = time.time()

random_make_num(randList3)

print("Random input list_sort before: ",randList3)
insertion_sort(randList3)
print("Random input list_ sort after: ",randList3)
print("For Loop test number of times: ",countNum+1)
randList3_countNum = countNum+1
countNum = 0
print("Inner While Loop test number of times: ",countLoopNum+rand_num-1)
randList3_countLoopNum = countLoopNum+rand_num-1
rand_num = 0
countLoopNum = 0

sec3 = time.time()-start3
```

```python
# 각 리스트에 대하여 실제 실행시간을 초, 시:분:초로 표시한다.
times1 = str(datetime.timedelta(seconds=sec1)).split(".")
times2 = str(datetime.timedelta(seconds=sec2)).split(".")
times3 = str(datetime.timedelta(seconds=sec3)).split(".")
times1 = times1[0]
times2 = times2[0]
times3 = times3[0]
print("Actual Running time_first_unit_sec_: ",sec1)
print("Actual Running time_first_unit_hour:minute:sec_: ",times1)
print("Actual Running time_second_unit_sec_: ",sec2)
print("Actual Running time_second_unit_hour:minute:sec_: ",times2)
print("Actual Running time_third_unit_sec_: ",sec3)
print("Actual Running time_third_unit_hour:minute:sec_: ",times3)
```

```python
# 세 가지 리스트에 대하여 각 요소의 개수의 평균, While Loop test의 평균 및 실제 실행시간의 평균을 구한다.
avg_count_num = (randList1_countNum + randList2_countNum + randList3_countNum) // 3
avg_count_loop_num = (randList1_countLoopNum + randList2_countLoopNum + randList3_countLoopNum) // 3
print("average n: ",avg_count_num)
print("Inner While Loop test number of times_average: ",avg_count_loop_num)


avg_sec = (sec1+sec2+sec3) / 3
times_avg = str(datetime.timedelta(seconds=avg_sec)).split(".")
times_avg = times_avg[0]
print("Actual Running time_avg_unit_sec_: ", avg_sec)
print("Actual Running time_third_unit_hour:minute:sec_: ",times_avg)
```

```python
x = [randList1_countNum, randList2_countNum, randList3_countNum]
y = [randList1_countLoopNum, randList2_countLoopNum, randList3_countLoopNum]
x.sort()
y.sort()

plt.plot(x,y,'bo-.')
plt.xlabel('n')
plt.ylabel('While loop test number of times')
plt.title('insertion_sort_random')
plt.show()
```

```python
x = [randList1_countNum, randList2_countNum, randList3_countNum]
y = [sec1,sec2,sec3]
x.sort()
y.sort()

plt.plot(x,y,'bo-.')
plt.xlabel('n')
plt.ylabel('Actual Running time')
plt.title('insertion_sort_random')
plt.show()
```
